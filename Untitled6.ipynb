```python
# Federated Learning with FedAvg Algorithm + Gradient Implementation + Visualizations

import numpy as np
import matplotlib.pyplot as plt
import torch
from torch import nn
from torch.utils.data import DataLoader, TensorDataset

# --- Generate Synthetic Data ---
def generate_balanced_synthetic_data(num_clients=5, num_samples=100, dim=2):
    datasets = []
    for _ in range(num_clients):
        while True:
            mean = np.random.rand(dim) * 10
            cov = np.eye(dim)
            X = np.random.multivariate_normal(mean, cov, num_samples)
            y = (X[:, 0] + X[:, 1] > 10).astype(int)
            if len(np.unique(y)) > 1:
                datasets.append((X, y))
                break
    return datasets

# --- PyTorch Logistic Regression Model ---
class LogisticModel(nn.Module):
    def __init__(self, input_dim):
        super(LogisticModel, self).__init__()
        self.linear = nn.Linear(input_dim, 1)

    def forward(self, x):
        return torch.sigmoid(self.linear(x))

# --- Train Function (1 client) ---
def train_local_model(X, y, model, epochs=5, lr=0.01):
    model.train()
    criterion = nn.BCELoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=lr)
    dataset = TensorDataset(torch.tensor(X, dtype=torch.float32), torch.tensor(y, dtype=torch.float32).view(-1, 1))
    loader = DataLoader(dataset, batch_size=32, shuffle=True)

    for _ in range(epochs):
        for xb, yb in loader:
            preds = model(xb)
            loss = criterion(preds, yb)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
    return model

# --- FedAvg Aggregation ---
def average_models(models):
    new_model = LogisticModel(models[0].linear.in_features)
    with torch.no_grad():
        weights = [m.linear.weight.data for m in models]
        biases = [m.linear.bias.data for m in models]
        new_model.linear.weight.data = torch.mean(torch.stack(weights), dim=0)
        new_model.linear.bias.data = torch.mean(torch.stack(biases), dim=0)
    return new_model

# --- FedAvg Loop ---
def simulate_fedavg_with_gradients(datasets, num_rounds=10, local_epochs=5):
    input_dim = datasets[0][0].shape[1]
    global_model = LogisticModel(input_dim)
    losses = []
    criterion = nn.BCELoss()

    for rnd in range(num_rounds):
        local_models = []
        for X, y in datasets:
            model = LogisticModel(input_dim)
            model.load_state_dict(global_model.state_dict())
            model = train_local_model(X, y, model, epochs=local_epochs)
            local_models.append(model)

        global_model = average_models(local_models)

        # Evaluate
        global_model.eval()
        with torch.no_grad():
            all_X = torch.tensor(np.vstack([d[0] for d in datasets]), dtype=torch.float32)
            all_y = torch.tensor(np.hstack([d[1] for d in datasets]), dtype=torch.float32).view(-1, 1)
            preds = global_model(all_X)
            loss = criterion(preds, all_y).item()
            losses.append(loss)
            print(f"[Round {rnd+1}] Loss: {loss:.5f}")

    return global_model, losses

# --- Plot Loss ---
def plot_fedavg_loss(losses):
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, len(losses)+1), losses, marker='o', linewidth=2, color='navy')
    plt.title("Federated Averaging with Local Gradient Descent")
    plt.xlabel("Federated Rounds")
    plt.ylabel("Loss")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# --- Visualize Client Data ---
def plot_client_data(datasets):
    plt.figure(figsize=(8, 6))
    colors = ['red', 'blue', 'green', 'orange', 'purple']
    for i, (X, y) in enumerate(datasets):
        plt.scatter(X[:, 0], X[:, 1], alpha=0.5, c=colors[i % len(colors)], label=f'Client {i+1}')
    plt.axline((0, 10), slope=-1, color='black', linestyle='--', label='True Boundary')
    plt.title("Synthetic Data Across Clients")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# --- Simulated Epoch Comparison (for visualization) ---
def simulate_loss_varying_epochs(num_rounds=50, client_epochs=1):
    np.random.seed(42)
    base_loss = np.exp(-np.linspace(0, 3, num_rounds))
    noise = np.random.normal(0, 0.05 / np.sqrt(client_epochs), num_rounds)
    return np.clip(base_loss + noise, 0, None)

def compare_epochs_loss(epoch_list, num_rounds=50):
    plt.figure(figsize=(10, 6))
    for ep in epoch_list:
        losses = simulate_loss_varying_epochs(num_rounds=num_rounds, client_epochs=ep)
        plt.plot(range(1, len(losses)+1), losses, label=f"{ep} Epochs", linewidth=2)
    plt.title("Simulated Loss Decay Across Epochs")
    plt.xlabel("Federated Rounds")
    plt.ylabel("Simulated Log Loss")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# --- Run the Simulation ---
if __name__ == '__main__':
    synthetic_datasets = generate_balanced_synthetic_data(num_clients=5, num_samples=100, dim=2)
    plot_client_data(synthetic_datasets)

    global_model, losses = simulate_fedavg_with_gradients(synthetic_datasets, num_rounds=10, local_epochs=5)
    plot_fedavg_loss(losses)

    compare_epochs_loss(epoch_list=[1, 3, 5, 10], num_rounds=50)
```
